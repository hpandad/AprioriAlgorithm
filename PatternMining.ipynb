```python
def freq_pattern_mining(input):
  import pandas as pd
  from itertools import combinations
  from operator import itemgetter
  i = 0
  minsup = 0
  lines = []
  with open(input, "r") as f:
    vals = []
    valsandsup = []
    maxchars = 0
    for item in f:
      lines.append(item)
      if i == 0:
        minsup = int(item[0])
      else:
        tempchar = 0
        for c in item:
          if c.isalpha():
            tempchar += 1
            if c in vals:
              index = vals.index(c)
              valsandsup[index][1] += 1
            else:
              vals.append(c)
              valsandsup.append([str(c),1])
        if tempchar > maxchars:
          maxchars = tempchar
      i += 1
    temp = 2
    table = []
    valsandsup.sort()
    while temp <= maxchars:
      j = 0
      listlen = len(valsandsup)
      newvals = ''
      while j < listlen:
        if len(valsandsup) == 0:
          break;
        if valsandsup[j][1] < minsup:
            # vals.remove(valsandsup[j][0])
            valsandsup.pop(j)
            j = j-1
            listlen -= 1
        else :
          for c in valsandsup[j][0]:
            if c not in newvals:
              newvals = newvals + c
        j+=1

      table.append(valsandsup)
      valsandsup = []
      if temp > len(newvals):
        break
      for comb in list(combinations(newvals, temp)):
          count = 0
          for x in lines:
            contains = True
            for c in comb:
              if c not in x:
                contains = False
            if contains:
              count+=1
          valsandsup.append([comb, count])
          count = 0
      j += 1
      temp += 1
  # print(table)
  frequent_patterns = []
  table3 = []
  row1 = True
  for temp in table:
    for t in temp:
      if row1:
        frequent_patterns.append([(t[1]), t[0]])
        table3.append([t[0], (t[1])])
        # table2.append(str(t[1]) + ": " + t[0])
      else:
        x = ""
        for z in t[0]:
          x = x + z
        frequent_patterns.append([(t[1]), x])
        table3.append([x, (t[1])])
        # table2.append(str(t[1]) + ": " + x)
    row1 = False
  table3 = sorted(table3, key=len)
  table3.reverse()
  # print(table3)
  # frequent_patterns.sort(reverse = True)
  # print(table2)
  func = lambda x: (-x[0], x[1])
  frequent_patterns = sorted(frequent_patterns, key = func)
  for t in frequent_patterns:
    print(str(t[0]) + ": " + t[1])
  closed_patterns = []
  for t in table3:
    should_add = True
    t_supp = t[1]
    for i in closed_patterns:
      if t_supp == i[1]:
        # print(t[0] + 'vs' + i[0])
        contains_all = True
        for c in t[0]:
          if c not in i[0]:
            contains_all = False
        if contains_all:
          should_add = False
          
    if should_add:
      closed_patterns.append(t)
  print("\n")
  new_closed = []
  for t in closed_patterns:
    new_closed.append([t[1], t[0]])
    # print(t[1] + ": " + t[0])
  new_closed.sort(reverse = True)
  func = lambda x: (-x[0], x[1])
  closed_patterns = sorted(closed_patterns, key = func)
  for i in new_closed:
    print(str(i[0]) + ": " + i[1])
  table4 = table3.copy()
  max_patterns = []
  for t in table4:
    should_add = True
    t_supp = t[1]
    for i in max_patterns:
        # print(t[0] + 'vs' + i[0])
        contains_all = True
        for c in t[0]:
          if c not in i[0]:
            contains_all = False
        if contains_all:
          should_add = False

          
    if should_add:
      max_patterns.append(t)
  new_max = []
  for t in max_patterns:
    new_max.append([t[1], t[0]])
    # print(t[1] + ": " + t[0])
  new_max.sort(reverse = True)
  print("\n")
  count = 0
  
  for i in new_max:
    count+=1
    print(str(i[0]) + ": " + i[1])
  print(count)
  
```
